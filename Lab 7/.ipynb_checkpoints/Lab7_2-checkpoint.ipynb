```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from mpl_toolkits.mplot3d import axes3d, Axes3D
path = './data'
```

```python
import os
img_list = os.listdir(path)
img_list
```

```python
f = np.zeros((256**2, 11))
for i, p in enumerate(img_list):
    img = plt.imread(os.path.join(path, p))
    f[:,i] = img.flatten()
```

```python
f[:,0].shape
```

```python
mu = np.mean(f, axis=0)
mu.shape
```

```python
Y = f - mu
Y.shape
```

```python
S = Y.T @ Y
```

```python
eig, V = np.linalg.eigh(S)
# change to descending order
eig = eig[::-1]
V = V[:,::-1]
eig
```

```python
u, s, vh = np.linalg.svd(Y)
u.shape
```

```python
Y_new = u[:,0:3].T @ Y
Y_new.shape
```

```python
fig = plt.figure(figsize=(5,15))
ax = Axes3D(fig)
ax.set_aspect("equal")

# Plot Points
for row in Y_new.T:
    ax.scatter(row[0], row[1], row[2], color="b", s=50, marker='*')
plt.show()
```
