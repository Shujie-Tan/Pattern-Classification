### EM Algorithm and Mixture of Densities (Gaussians) for Image Segmentation

Segmentation of MRI images using mixed Gaussian models for
provided data sets – you need to determine the number of Gaussian
models, mean, and variance for each Gaussian model, also use color
for different categories.

```python
def gausspdf1D(x, mu, sigma):
    mu = np.array(mu)
    sigma = np.array(sigma)
    # x is a list
    A = (2 * np.pi)**(1/2) * sigma # scalar
    B = np.multiply((x-mu),(x-mu)) / sigma**2
    p = 1 / A * np.exp(- B / 2)
    return p
```

```python
# define Gauss pdf in sympy manner
import sympy as sp
def gausspdfsp(x, mu, sigma):
    mu = np.array(mu)
    sigma = np.array(sigma)
    # x is a list
    A = (2 * np.pi)**(1/2) * sigma # scalar
    B = np.multiply((x-mu),(x-mu)) / sigma**2
    p = 1 / A * sp.exp(- B / 2)
    return p

x, y  = sp.symbols('x y')
```

```python
def GMM_EM(y, K):
    # Initialization of parameters
    y = np.squeeze(y)
    N = len(y)
    mu = np.random.randint(255, size=K)
    sigma = np.random.randint(50,size=K)
    alpha = np.random.rand(K)
    alpha = alpha / alpha.sum()
    threshold,prob,step = 0.01,0,0
    tmp = np.zeros((N,K))
    while True:
        step +=1
        for k in range(K):  # E-step
            tmp[:,k] = alpha[k] * gausspdf1D(y, mu[k], sigma[k])
        tmp_sum = np.sum(tmp, axis=1)
        new_prob = np.sum(np.log(tmp_sum))

        if np.abs(new_prob - prob) > threshold:
            prob = new_prob
        else:
            break 
        tmp_sum_reciprocal = 1 / tmp_sum
        tmp_sum_reciprocal = tmp_sum_reciprocal[:,np.newaxis]
        tmp_sum_reciprocal = np.repeat(tmp_sum_reciprocal, K,axis=1)
        gamma = tmp * tmp_sum_reciprocal # equal above
        gamma_sum = np.sum(gamma,axis=0)                # M step
        old_mu, old_sigma, old_alpha = mu, sigma, alpha
        mu = y @ gamma / gamma_sum
        sigma = np.zeros((K))
        for k in range(K):
            sigma[k] = ((y-mu[k]) * (y-mu[k])) @ gamma[:,k]
        sigma = sigma / gamma_sum
        sigma = np.sqrt(sigma)
        alpha = np.sum(gamma,axis=0) / N
        if step > 400:
            break
    return mu, sigma, alpha #, gamma
```

```python
K = 4 
mu, sigma, alpha = GMM_EM(test,K)
```

```python
x, y = sp.symbols('x y')
```

```python
def sol(z, low, high):
    for i in range(int(low), int(high)):
        if z.subs(x, i) * z.subs(x, i+1) <= 0:
            break
    return i
```

```python
i, j = 0, 1
z = alpha[i] * gausspdfsp(x, mu[i], sigma[i]) - \
      alpha[j] * gausspdfsp(x, mu[j], sigma[j])
seg1 = sol(z, mu[i], mu[j])
seg1
```

```python
i, j = 1, 2
z = alpha[i] * gausspdfsp(x, mu[i], sigma[i]) - \
      alpha[j] * gausspdfsp(x, mu[j], sigma[j])
seg2 = sol(z, mu[i], mu[j])
seg2
```

```python
i, j = 2, 3
z = alpha[i] * gausspdfsp(x, mu[i], sigma[i]) - \
      alpha[j] * gausspdfsp(x, mu[j], sigma[j])
seg3 = sol(z, mu[i], mu[j])
seg3
```

```python
seg1, seg2, seg3
```

```python
# segmentation
MRI1_seg = np.copy(MRI1)
MRI1_seg = MRI1_seg[:,:,np.newaxis]
MRI1_seg = np.repeat(MRI1_seg, 3, axis=2)
```

```python
def rgb_seg(img, seg_img, low, high, color):
    assert img.ndim == 2
    assert len(color) == 3
    index = np.multiply(img >= low, img<high)
    for i in range(3):
        seg_img[index,i] = color[i]
    return seg_img
```

```python
rgb_seg(MRI1,MRI1_seg, 0, seg1, [0,0,0])
rgb_seg(MRI1,MRI1_seg, seg1, seg2, [0,255,0])
rgb_seg(MRI1,MRI1_seg, seg2, seg3, [255,0,0])
rgb_seg(MRI1,MRI1_seg, seg3, 256, [0,0,255])
plt.imshow(MRI1_seg)
```
